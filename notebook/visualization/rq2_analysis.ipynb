```python
result_path = "/home/fmg2/v-thanh/Code"
without_db = "results/TSBA/iAWE/without_bd/bb_resnet_exp_0-t_3-clf_fcn/attack_metrics_resnet_to_fcn.csv"
with_db = "results/TSBA/iAWE/with_bd/bb_resnet_exp_0-t_3-clf_fcn_old/attack_metrics_resnet_to_fcn.csv"

without_db_df = pd.read_csv(f"{result_path}/{without_db}")
with_db_df = pd.read_csv(f"{result_path}/{with_db}")

print(without_db_df.columns)
print(with_db_df.columns)

without_db_df = without_db_df[without_db_df["Attack Epoch"] == 2]
with_db_df = with_db_df[with_db_df["Attack Epoch"] == 2]

print(len(without_db_df), len(with_db_df))
```

```python
target_asr_without_db = without_db_df["Target ASR"].values[:40]
target_asr_with_db = with_db_df["Target ASR"].values[:40]
```

```python
plt.figure(figsize=(14, 8))

# Create the plot with better styling
plt.plot(range(1, len(target_asr_without_db) + 1), target_asr_without_db, 
         label='Training without poisoned data', marker='o', markersize=4, linewidth=2, 
         color='#2E86AB', alpha=0.8)
plt.plot(range(1, len(target_asr_with_db) + 1), target_asr_with_db, 
         label='Training with poisoned data', marker='s', markersize=4, linewidth=2, 
         color='#A23B72', alpha=0.8)

# Improve labels and title
plt.xlabel('Epochs', fontsize=23, fontweight='bold')
plt.ylabel('Target ASR', fontsize=23, fontweight='bold')

# Enhance legend
plt.legend(fontsize=25, frameon=True, fancybox=True, shadow=True)

# Improve grid
plt.grid(True, alpha=0.3, linestyle='--')

# Set y-axis limits for better visualization
plt.ylim(0.4, 1.1)

# Add minor ticks
plt.minorticks_on()

# Increase font size for tick labels
plt.tick_params(axis='both', which='major', labelsize=20)

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
