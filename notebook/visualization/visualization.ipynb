```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
root_dir = "/home/fmg2/v-thanh/Code/results/TSBA/iAWE/others"
gen1_file = f"{root_dir}/generator_epoch_1_acc.csv"
gen2_file = f"{root_dir}/generator_epoch_2_acc.csv"
gen3_file = f"{root_dir}/generator_epoch_3_acc.csv"
gen10_file = f"{root_dir}/generator_epoch_10_acc.csv"
gen20_file = f"{root_dir}/generator_epoch_20_acc.csv"
gen30_file = f"{root_dir}/generator_epoch_30_acc.csv"
gen40_file = f"{root_dir}/generator_epoch_40_acc.csv"
gen50_file = f"{root_dir}/generator_epoch_50_acc.csv"

# Load the CSV files
gen1_df = pd.read_csv(gen1_file)
gen2_df = pd.read_csv(gen2_file)
gen3_df = pd.read_csv(gen3_file)
gen10_df = pd.read_csv(gen10_file)
gen20_df = pd.read_csv(gen20_file)
gen30_df = pd.read_csv(gen30_file)
gen40_df = pd.read_csv(gen40_file)
gen50_df = pd.read_csv(gen50_file)
```

```python
def plot_generator_losses_improved(gen_df=gen3_df):
    # Set up the figure with better styling
    plt.style.use('seaborn-v0_8-whitegrid')  # Use seaborn style for better aesthetics
    fig, ax = plt.subplots(figsize=(12, 8))  # Reduced width since we only have one line
    
    # Plot gen1_df only
    final_asr = gen_df['Value'].iloc[-1]
    label_with_asr = f"Generator Epoch 1 (ASR: {final_asr:.3f})"
    
    ax.plot(gen_df['Step'], gen_df['Value'], 
            label=label_with_asr,
            linewidth=2.5, 
            color='#1f77b4',  # Default blue color
            alpha=0.8,
            marker='o',
            markersize=4,
            markevery=max(1, len(gen_df)//10))  # Show markers every 10% of points
    
    # Enhance the plot appearance with larger font sizes
    ax.set_xlabel('Training Epochs', fontsize=20, fontweight='bold')
    ax.set_ylabel('Attack Success Rate', fontsize=20, fontweight='bold')
    ax.set_title('Training progress of First epoch Generator', 
                fontsize=22, fontweight='bold', pad=20)
    
    # Add legend
    legend = ax.legend(fontsize=14,
                      frameon=True, 
                      fancybox=True)
    
    legend.get_frame().set_facecolor('white')
    legend.get_frame().set_alpha(0.9)
    
    # Enhance grid
    ax.grid(True, alpha=0.3, linestyle='-', linewidth=0.5)
    ax.set_axisbelow(True)
    
    # Improve tick labels with larger font size
    ax.tick_params(axis='both', which='major', labelsize=14)
    ax.tick_params(axis='both', which='minor', labelsize=14)
    
    # Set y-axis to start from 0 and add some padding
    y_min = gen_df['Value'].min()
    y_max = gen_df['Value'].max()
    ax.set_ylim(max(0, y_min - 0.05), y_max + 0.05)
    
    # Add horizontal reference lines for better interpretation
    ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.5, linewidth=1)
    
    # Add text annotations for reference lines with larger font
    ax.text(ax.get_xlim()[1] * 0.98, 0.5, 'Random Guess', 
            verticalalignment='bottom', horizontalalignment='right',
            fontsize=16, alpha=0.7)

    plt.tight_layout()
    plt.show()
    
    return fig, ax

plot_generator_losses_improved(gen3_df)
```
